```python
import numpy as np
import json
```

```python
load_path = "../data/compressed/event_metadata.json"
with open(load_path) as fp:
    event_meta = json.load(fp)
```

```python
richters = []
for event in event_meta:
    richters.append(event['mag'])
print(len(richters))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()

plt.figure(figsize=(8, 8))
plt.hist(richters, bins=50)
plt.xlabel("Richter reading"); plt.title("Distribution of Richter magnitudes")
plt.savefig('./figures/richters.png')
plt.show()
```

```python
np.save("./misc/richters.npy", np.array(richters))
```
